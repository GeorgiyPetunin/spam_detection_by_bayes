```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import CategoricalNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.naive_bayes import ComplementNB
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfTransformer
```

```python
def cmf(y_test,y_pred):
    cm = confusion_matrix(y_test, y_pred)

    plt.figure(figsize=(3, 3))
    sns.heatmap(cm, annot=True, fmt='d')
    plt.title('Матриця помилок')
    plt.xlabel('Прогнозована мітка')
    plt.ylabel('Справжня мітка')
    plt.show()
```

```python
def accuraciesGraphic(acc : list[float], clas : list[str], name : str, x : float = 0):
    accuracies = acc
    classifiers = clas

    plt.figure(figsize = (7, 3))
    plt.bar(classifiers, accuracies)
    plt.xlabel('Classifiers')
    plt.ylabel(name)
    plt.title('Distribution of Classifier ' + name)
    plt.ylim(x, 1)

    plt.show()
```

```python
df = pd.read_csv('bd.csv')
```

```python
x_train, x_test, y_train, y_test = train_test_split(df['part'], df['spam'], test_size = 0.1, shuffle = 1)
```

```python
cv = CountVectorizer(lowercase = True)
x_train_count = cv.fit_transform(x_train.values)
x_test_count = cv.transform(x_test.values)

x_train_count = x_train_count.toarray()
x_test_count = x_test_count.toarray()
```

```python
clf1 = GaussianNB()
clf1.fit(x_train_count, y_train)

y_pred = clf1.predict(x_test_count)

clf1a=clf1.score(x_test_count, y_test)
clf1p=precision_score(y_test, y_pred)
clf1r=recall_score(y_test, y_pred)
clf1f=recall_score(y_test, y_pred) * 2 * precision_score(y_test, y_pred) / (precision_score(y_test, y_pred) + recall_score(y_test, y_pred))

print('Accuracy of training set = ' + format(clf1.score(x_train_count, y_train)))
print('Accuracy of test set = ' + format(clf1a))
print('Precision score of test set = ' + format(clf1p))
print('Recall score of test set = ' + format(clf1r))
print('F_weight score of test set = ' + format(clf1f))

cmf(y_test,y_pred)
```

```python
clf2 = BernoulliNB()
clf2.fit(x_train_count, y_train)

y_pred = clf2.predict(x_test_count)

clf2a=clf2.score(x_test_count, y_test)
clf2p=precision_score(y_test, y_pred)
clf2r=recall_score(y_test, y_pred)
clf2f=recall_score(y_test, y_pred) * 2 * precision_score(y_test, y_pred) / (precision_score(y_test, y_pred) + recall_score(y_test, y_pred))

print('Accuracy of training set = ' + format(clf2.score(x_train_count, y_train)))
print('Accuracy of test set = ' + format(clf2a))
print('Precision score of test set = ' + format(clf2p))
print('Recall score of test set = ' + format(clf2r))
print('F_weight score of test set = ' + format(clf2f))

cmf(y_test,y_pred)
```

```python
clf4 = MultinomialNB()
clf4.fit(x_train_count, y_train)

y_pred = clf4.predict(x_test_count)

clf4a=clf4.score(x_test_count, y_test)
clf4p=precision_score(y_test, y_pred)
clf4r=recall_score(y_test, y_pred)
clf4f=recall_score(y_test, y_pred) * 2 * precision_score(y_test, y_pred) / (precision_score(y_test, y_pred) + recall_score(y_test, y_pred))

print('Accuracy of training set = ' + format(clf4.score(x_train_count, y_train)))
print('Accuracy of test set = ' + format(clf4a))
print('Precision score of test set = ' + format(clf4p))
print('Recall score of test set = ' + format(clf4r))
print('F_weight score of test set = ' + format(clf4f))

cmf(y_test,y_pred)
```

```python
clf6 = ComplementNB()
clf6.fit(x_train_count, y_train)

y_pred = clf6.predict(x_test_count)

clf6a=clf6.score(x_test_count, y_test)
clf6p=precision_score(y_test, y_pred)
clf6r=recall_score(y_test, y_pred)
clf6f=recall_score(y_test, y_pred) * 2 * precision_score(y_test, y_pred) / (precision_score(y_test, y_pred) + recall_score(y_test, y_pred))

print('Accuracy of training set = ' + format(clf6.score(x_train_count, y_train)))
print('Accuracy of test set = ' + format(clf6a))
print('Precision score of test set = ' + format(clf6p))
print('Recall score of test set = ' + format(clf6r))
print('F_weight score of test set = ' + format(clf6f))

cmf(y_test,y_pred)
```

```python
pipeline = Pipeline([
    ('vect', CountVectorizer(lowercase = True)),
    ('tfidf', TfidfTransformer(use_idf = True, smooth_idf = True)),
    ('clf', MultinomialNB(alpha = 1))
])

mod1 = pipeline.fit(x_train, y_train)

y_pred = mod1.predict(x_test)

moda1=mod1.score(x_test, y_test)
modp1=precision_score(y_test, y_pred)
modr1=recall_score(y_test, y_pred)
modf1=recall_score(y_test, y_pred) * 2 * precision_score(y_test, y_pred) / (precision_score(y_test, y_pred) + recall_score(y_test, y_pred))

print('Accuracy of training set = ' + format(mod1.score(x_train, y_train)))
print('Accuracy of test set = ' + format(moda1))
print('Precision score of test set = ' + format(modp1))
print('Recall score of test set = ' + format(modr1))
print('F_weight score of test set = ' + format(modf1))

cmf(y_test,y_pred)
```

```python
accuracyRate = [clf1a, clf2a, clf4a, clf6a, moda1]
preccisionRate = [clf1p, clf2p, clf4p, clf6p, modp1]
recallRate = [clf1r, clf2r, clf4r, clf6r, modr1]
f1Rate = [clf1f, clf2f, clf4f, clf6f, modf1]
algoNames = ['Gaussian', 'Bernoulli', 'Multinomial', 'Complement', 'Multinomial_UP']

accuraciesGraphic(accuracyRate, algoNames, 'accuracies', 0.89)
accuraciesGraphic(preccisionRate, algoNames, 'preccision', 0.8)
accuraciesGraphic(recallRate, algoNames, 'recall', 0.7)
accuraciesGraphic(f1Rate, algoNames, 'f_weight', 0.8)
```
